# variable length argument

*sometimes we pass variable number of argument*

```python
# def f1(*n):
```

```python
def sum(*n):
    total=0
    for i in n:
        total=total+i
    print(total)
sum()
```

```python
sum(5,50,10)
```

```python
# write the code for finding facctorial
def factorial(n) :
    result=1
    for i in range(1,n+1):
        result= result*i
    return result
```

```python
factorial(4)
```

```python
# recursive function
def fact(n):
    if n==0:
        result=1
    else:
        result=n*fact(n-1)
    return result
```

```python
fact(5)
```

# filter function

*we can use filter() to filter values from the given sequence based on the given condition*

```python
# to find the even numbers from the list by using filter ()
```

```python
def isEven(x):
    if x%2==0: 
        return True 
    else :
        return False
L=[10,20,4,5,9,8,7]
L1=list(filter (isEven,L))
L1
```

```python
l=[5,6,7,8,9,2,3]
l1=list(filter(lambda x:x%2==0,l))
l1
```

```python
l=[5,6,7,8,9,2,3]
l1=list(filter(lambda x:x%2!=0,l))
l1
```

# map ()

*for every elements present in the given sequence apply some functionality and generate new element with the required modification we use map()*

```python
#  map(function,sequence)
```

```python
l=[12,10,30,40,50,60]
def doubleit(x):
    return 2*x
l1=list(map(doubleit,l))
l1
```

# reduce()

*it reduces sequence of elements into a single element by applying specific functions* .
*it is present in functional tools module and hence we should write import statements*

```python
# reduce(function,sequence)
```

```python
from functools import *
l=[10,20,30,40]
result=reduce(lambda x,y:x+y,l)
print(result)
```
