# SET

*set represents a group of unique elements*

# features of a set

- duplicates are not allowed
- insertion order is not preserved
- we can sort the elements
- indexing and slicing are not allowed
- heterogenuous object are allowed

### defining a set : set is defined using {}

```python
s={10,20,30}
type(s)
```

```python
#  duplicates are not allowed
s={10,20,10,20,10}
```

```python
s
```

```python
#  list to set
l=[10,20,30,50,40,60]
set(l)
```

```python
#  using range
s=set(range(10))
```

```python
s
```

# mathematical operators of set

- union() : combines elements present in both set
- intersection() : gives common elements only

### **union()** - |

```python
a={1,2,3}
a1={4,5,6}
a.union(a1)
```

```python
r={5,6,8}
r1={7,8,9}
r|r1
```

### **intersection()** :&

```python
d={78,98,45,34,76}
d1={98,78,89,56,45}
d&d1
```

```python
d.intersection(d1)
```
