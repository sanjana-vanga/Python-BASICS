# set data type

- when we want to represent a group of values without duplicates where order is not important then we should go for set data type
- insertion order is not preserved
- duplicates are not allowed
- heterogenous objects are allowed
- index concept is not applicable in set  data type
- it is mutable and growable in nature
- we enclose in {} .

```python
s={10,20,22.5,5+9j,"hari",True,10}
```

```python
s
```

```python
#s[0]
```

```python
s.add(55)
```

```python
s
```

```python
s.remove(True)
```

```python
s
```
