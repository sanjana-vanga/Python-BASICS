# if -elif -else

```python
brand=input("what is your favourite brand ?")
if brand=="kf" :
    print("this is a legend brand")
elif brand=="old monk":
    print("this is the brand of ice")
elif brand=="signature":
    print("this is my brand")
else:
    print("no discount")
```

###### q2 . finding out the biggest/smallest  number in 3 inputs

```python
# finding the biggest/smallest among 3 values
a,b,c=(input("Enter 3 values:")).split()
if a>b and a>c:
    print("The biggest among three is {0}".format(a))
elif b>a and b>c:
    print("The biggest among three is {0}".format(b))
else:
    print("The biggest among three is {0}".format(c))
```
