# ***while loop***

```python
x=1
while x<=10:
    print(x)
    x+=1
```

```python
n=int(input("enter the first number"))
sum=0
i=1
while i<=n:
    sum+=i
    i+=1
    print("the sum of first",n,"numbers is",sum)
```

*loop within the loop*

# *Transfer Statements*

### break

```python
for i in range(10):
    if i==7:
        print("process is enough")
        break
    print(i)
```

```python
# break in a list
cart=(100,200,300,400,500,600)
for item in cart:
    if item>500:
        print("you get discount")
        break
    print(item)
```

# *continue*

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
cart=(100,200,300,400,500,600,700,120,103)
for item in cart:
    if item>500:
        print("item unavailable")
        continue
    print(item)
```

# *pass*

- pass is a keyword in python
- In python if a block is required which won't do anything then we can define that empty block with pass keyword

```python
for i in range(10):
    if i%2==0:
         print(i)
    else:
        pass
```
