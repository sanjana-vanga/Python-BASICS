**in string literals we can use escape character to associate a special character**

```python
#\n give new line
da="data\nanalytics"
```

```python
print(da)
```

```python
#\t for one tab distance
```

```python
h="Graphics\tdesign"
```

```python
# \b for backspace
j="mani\b\b\b\bpopulation"
```

```python
print(j)
```

```python
# \v for vertical line
j="mani\vpopulation"
print(j)
```

```python
j="mani\rpopulation"
print(j)
```

```python
j="mani\fpopulation"
print(j)
```

```python
j="mani\'population"
print(j)
```

```python
j="mani\\population"
print(j)
```
