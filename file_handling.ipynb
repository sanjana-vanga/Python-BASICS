# file handling

1. text file ---> abc.text
2. binary file ---> image,audio,video

# opening file in python

```python
f=open("abc.txt","w")
print("file name",f.name)
print("file name",f.mode)
print("is it readable",f.readable())
print("is it readable",f.writable())
f.close()
```

# writing text in the file


```python
f=open("abc.txt","w")
f.write("rakshith\n")
f.write("kalyan")
print("written")
f.close()
```
