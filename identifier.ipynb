# rules to define identifiers in python

1. alphabet are includes both upper and lower case (A-Z ,a-z)
2. digits includes (0-9)
3. underscore ( _ ) allowed
4. $ symbol not allowed
5. identifiers should start with character not with number
6. identifiers are case sensitive
7. we can't use reserve words as identifiers
