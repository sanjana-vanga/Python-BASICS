# How to take multiple i/p at a time

```python
# split() ->splitting the vals.It'll take space as a seperator by default.
a,b=[int(x) for x in input ("enter two values") .split()]
print("product:",a*b)
```

```python
a,b,c=[float(x) for x in input("Enter three values:").split()]
print("sum of three:",a+b+c)
```

# formated string

```python
name="pavan"
salary=50000
friend="sunny"
print("hello {0} your salary {1} and friend is {2}". format(name,salary,friend))
print("hello {x} your salary {y} and friend is {z}". format(x=name,y=salary,z= friend))
```
