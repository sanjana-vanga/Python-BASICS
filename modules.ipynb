# random module

# randint()

```python
from random import *
for i in range(10):
    print(randint(1,30))
```

# uniform()

```python
from random import *
for i in range(10):
    print(uniform(1,10))
```

# randrange([start,stop,step])

```python
from random import *
for i in range(10):
    print(randrange(1,30,2))
```

# choice()

```python
from random import *
list1=["sunny","dinnu","rahul","mani"]
for i in range(10):
    print(choice(list1))
```
