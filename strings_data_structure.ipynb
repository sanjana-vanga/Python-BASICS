# removing spaces from the string

there are 3 mothods to remove space
- 1. rstrip() ==> remove space right hand side
- 2. lstrip() ==> remove space left hand side
- 3. strip()  ==> remove space both side

```python
city=input("enter your city name")
scity=city.strip()
if scity=="hyderabad":
    print("hello hyderabadi...adaab")
elif scity=="warangal":
    print("hello warangal..namaskaram")
elif scity=="chennai":
    print("hello madrasi...vanakam")
else:
    print("enter valid city name")
```
